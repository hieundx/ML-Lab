```python
import numpy as np
import matplotlib.pyplot as plt

# Load data
data = np.loadtxt(
    './Student_Performance.csv',
    delimiter=',',
    skiprows=1,
    dtype=[
        ('Hours Studied', 'float'),
        ('Previous Scores', 'float'),
        ('Extracurricular Activities', 'U3'),
        ('Sleep Hours', 'float'),
        ('Sample Question Papers Practiced', 'float'),
        ('Performance Index', 'float'),
    ]
)
```

```python
# Convert data for machine learning
m = len(data)

perf_idx = 5

x_train = np.zeros((m, perf_idx))
y_train = np.zeros(m)
print(x_train.shape)
for i in range(m):
    entry = data[i]
    row = np.zeros(perf_idx)
    for j in range(perf_idx):
        val = entry[j]
        if val == 'Yes':
            val = 1
        elif val == 'No':
            val = 0
        row[j] = val
    x_train[i] = row
    y_train[i] = entry[perf_idx]


print(x_train[0:10])
print(y_train[0:10])
```

```python
# Visualize data
fig, axs = plt.subplots(2, 2)

x_hours_studied = x_train[:, 0]
axs[0, 0].scatter(x_hours_studied, y_train, s=1)
axs[0, 0].set_title('Hours studied')

x_prev_scores = x_train[:, 1]
axs[0, 1].scatter(x_prev_scores, y_train, s=1)
axs[0, 1].set_title('Previous Scores')

x_sleep_hours = x_train[:, 3]
axs[1, 0].scatter(x_sleep_hours, y_train, s=1)
axs[1, 0].set_title('Sleep Hours')

x_questions_practiced = x_train[:, 4]
axs[1, 1].scatter(x_questions_practiced, y_train, s=1)
```

```python
# OPTIONAL: normalize data
mu = np.mean(x_train)
sigma = np.std(x_train)

x_norm = (x_train - mu)/sigma

# we'll train our model using x_train for now
# and swap to normalized data later for comparison
X = x_train
```

```python
# Define initial weights and biases

w = np.array(np.random.uniform(1, 21, size=5))
b = np.random.uniform(1, 21)
```

```python
def calculate_perf_index(x, w, b):
    return np.dot(w, x) + b

# test run the function
# calculate_perf_index([ 7., 99.,  1.,  9., 1.], w, b)
```

```python
def calculate_cost(X, y, w, b):
    cost = 0
    m = X.shape[0]
    
    for i in range(m):
        y_hat = calculate_perf_index(X[i], w, b)
        y_i = y[i]
        
        cost += (y_hat - y_i)**2
    cost /= (2*m)
    return cost

# test run cost function
# calculate_cost(X[:20], y_train[:20], w, b)
```

```python
def calculate_gradients(X, y, w, b):
    m, n = X.shape
    
    dj_dw = np.zeros(n)
    dj_db = 0

    for i in range(m):
        y_hat = calculate_perf_index(X[i], w, b)
        y_i = y[i]

        err = y_hat - y_i # scalar
        dj_dw += err * X[i]
        dj_db += err

    dj_dw /= m
    dj_db /= m

    return dj_dw, dj_db

# test run gradient calculator
calculate_gradients(X[:10], y_train[:10], w, b)
```

```python
def train(X, y, w, b, learning_rate, epsilon):
    m, n = X.shape
    num_iters = 100001

    costs = []

    for i in range(num_iters):
        dj_dw, dj_db = calculate_gradients(X, y, w, b)

        w -= learning_rate * dj_dw
        b -= learning_rate * dj_db

        cost = calculate_cost(X, y, w, b)
        costs.append(cost)

        if cost < epsilon:
            break
        
        if i % 1000 == 0:
            print('[', i, ']' ,' cost: ', costs[i])

    return costs, w, b
```

```python
w = np.array(np.random.uniform(1, 21, size=5))
b = np.random.uniform(1, 21)
alpha = 1e-3
epsilon = 16

cost_history, w_res, b_res = train(x_norm, y_train, w, b, alpha, epsilon)

print('Weights and bias found:', w_res, b_res, 'after', len(cost_history), 'iterations')
```

```python
# visualize learning process
fig, ax = plt.subplots()
ax.plot(np.arange(len(cost_history)), cost_history)
```

```python
# test first 5 predictions
m = x_train[:20].shape[0]

for i in range(m):
    prediction = calculate_perf_index(x_norm[i], w_res, b_res)
    target = y_train[i]

    print('{:.2f}  {:.2f}, {:.2f}'.format(prediction, target, abs(prediction - target)))
```

```python
# Evaluate the model
m = x_train.shape[0]
y_predict = []
for i in range(m):
    y_predict.append(calculate_perf_index(x_norm[i], w_res, b_res))

# Visualize data
fig, axs = plt.subplots(2, 2)

x_hours_studied = x_train[:, 0]
axs[0, 0].scatter(x_hours_studied, y_train, s=1)
axs[0, 0].scatter(x_hours_studied, y_predict, s=1)
axs[0, 0].set_title('Hours studied')

x_prev_scores = x_train[:, 1]
axs[0, 1].scatter(x_prev_scores, y_train, s=1)
axs[0, 1].scatter(x_prev_scores, y_predict, s=1)
axs[0, 1].set_title('Previous Scores')

x_sleep_hours = x_train[:, 3]
axs[1, 0].scatter(x_sleep_hours, y_train, s=1)
axs[1, 0].scatter(x_sleep_hours, y_predict, s=1)
axs[1, 0].set_title('Sleep Hours')

x_questions_practiced = x_train[:, 4]
axs[1, 1].scatter(x_questions_practiced, y_train, s=1)
axs[1, 1].scatter(x_questions_practiced, y_predict, s=1)
```

```python
# Scikit-Learn implementation
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
skx_norm = scaler.fit_transform(x_train)
```

```python
skx_norm[:5]
```

```python
sgdr = SGDRegressor(max_iter=10000)
sgdr.fit(skx_norm, y_train)
print(sgdr)
print(f"number of iterations completed: {sgdr.n_iter_}, number of weight updates: {sgdr.t_}")
```

```python
b_norm = sgdr.intercept_[0]
w_norm = sgdr.coef_
print(f"model parameters:                   w: {w_norm}, b:{b_norm}")
```

```python
# test first 5 predictions
m = x_train[:20].shape[0]

for i in range(m):
    prediction = calculate_perf_index(skx_norm[i], w_norm, b_norm)
    target = y_train[i]

    print('{:.2f}  {:.2f}, {:.2f}'.format(prediction, target, abs(prediction - target)))
```

```python
# Evaluate the model
m = x_train.shape[0]
y_predict_sgdr = []
for i in range(m):
    y_predict_sgdr.append(calculate_perf_index(x_norm[i], w_res, b_res))

# Visualize data
fig, axs = plt.subplots(2, 2)

x_hours_studied = x_train[:, 0]
axs[0, 0].scatter(x_hours_studied, y_train, s=1)
axs[0, 0].scatter(x_hours_studied, y_predict_sgdr, s=1)
axs[0, 0].set_title('Hours studied')

x_prev_scores = x_train[:, 1]
axs[0, 1].scatter(x_prev_scores, y_train, s=1)
axs[0, 1].scatter(x_prev_scores, y_predict_sgdr, s=1)
axs[0, 1].set_title('Previous Scores')

x_sleep_hours = x_train[:, 3]
axs[1, 0].scatter(x_sleep_hours, y_train, s=1)
axs[1, 0].scatter(x_sleep_hours, y_predict_sgdr, s=1)
axs[1, 0].set_title('Sleep Hours')

x_questions_practiced = x_train[:, 4]
axs[1, 1].scatter(x_questions_practiced, y_train, s=1)
axs[1, 1].scatter(x_questions_practiced, y_predict_sgdr, s=1)
```
